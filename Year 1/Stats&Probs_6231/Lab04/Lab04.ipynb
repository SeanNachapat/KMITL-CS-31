```python
import numpy as np
import pandas as pd
df= pd.read_csv('playtennis.csv')
df
```

```python
X=df.drop(['play'], axis=1)
y=df.play
```

## **Label**

```python
from sklearn.preprocessing import LabelEncoder
def labelEncode(data,columns):
  for i in columns:
    lb=LabelEncoder().fit_transform(data[i])
    data[i+'_'] = lb
```

```python
f_columns=['outlook', 'temp','humidity', 'windy']
labelEncode(X,f_columns)
X
```

```python
y_le=LabelEncoder()
y1=y_le.fit_transform(y)
y1
```

```python
X1=X[['outlook_', 'temp_','humidity_', 'windy_']]
X1
```

## **Model Construction**

```python
from sklearn.naive_bayes import CategoricalNB
model=CategoricalNB()
model.fit(X1,y1)
```

## **Prediction**

```python
new_input=pd.DataFrame([[2,0,1,1],[1,2,1,0]],columns=['outlook_', 'temp_','humidity_', 'windy_'])
```

```python
y_prob_pred = model.predict_proba(new_input)
y_prob_pred.round(3)
```

```python
y_new_predict=model.predict(new_input)
n=1
for i in y_new_predict:
  print( 'No' ,n, '=>: ',y_le.classes_[i])
  n=n+1
```
