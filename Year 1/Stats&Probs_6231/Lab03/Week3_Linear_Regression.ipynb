Predict the value of Y when X is {300,500,1000}

```python
newX=[[300],[500],[1000]]
y_predict=model.predict(newX)
y_predict
```

### **Model Evaluation**

Coefficient of determination (R2)

$R^2 = 1- \frac{RSS}{TSS}$ ,
$RSS = âˆ‘^n_{i=1}(y_i-f(x_i))^2$ ,
$TSS =âˆ‘^n_{i=1}(y_i-yÌ„)^2$

```python
model.score(x_TV,y)
```

Mean Absolute Error (MAE)

$MAE = \frac{âˆ‘^n_{i=1}|y_i-x_i|}{n}$

Mean Square Error (MSE)

$MSE = \frac{1}{n} âˆ‘^n_{i=1}(y_i-yÌ‚_i)^2$

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
y_predict1=model.predict(x_TV)
print('MAE ='

, mean_absolute_error(y,y_predict1))

print('MSE ='

, mean_squared_error(y,y_predict1))
```

# **Setup Polynomial Regression**

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = np.array([0, 1, 2, -1, -2]).reshape(-1,1)
y = np.array([1, 6, 17, 2, 9]).reshape(-1,1)
```

```python
poly_features=PolynomialFeatures(degree=2)
x_poly=poly_features.fit_transform(x)
model=LinearRegression()
model.fit(x_poly,y)
```

```python
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

# **Example Polynomial Regression | Salary**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df= pd.read_csv('salary.csv')
df
```

```python
x = df.iloc[:,1:2].values
y = df.iloc[:,2].values
x, y
```

```python
plt.scatter(x,y, color='red')
plt.ticklabel_format(style='plain')
plt.title('Salary vs Level')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## **Evaluate Degrees**

```python
degree=['Degree1', 'Degree2', 'Degree3', 'Degree4']
Predict=pd.DataFrame(index=degree).T #transpose of DataFrame
Rscore = []
```

```python
for k in range(1, 5):
  poly_features=PolynomialFeatures(degree=k)
  x_poly=poly_features.fit_transform(x)
  model=LinearRegression()
  model.fit(x_poly,y)
  p1=model.predict(x_poly)
  if(k==1):
    Predict.Degree1=p1
  elif(k==2):
    Predict.Degree2=p1
  elif(k==3):
    Predict.Degree3=p1
  else:
    Predict.Degree4=p1
  Rscore.append(model.score(x_poly,y))
```

```python
Predict
```

```python
Rscore
```

## **Model's Prediction Visualization**

```python
plt.figure(figsize=(10,6))
plt.ticklabel_format(style='plain')
plt.scatter(x,y,color='b')
plt.plot(x,Predict.Degree1,linewidth='3',color='g',linestyle='dotted',label='linear(degree 1)')
plt.plot(x,Predict.Degree2,linewidth='3',color='r',linestyle='dotted',label='degree 2')
plt.plot(x,Predict.Degree3,linewidth='3',color='yellow', linestyle='solid',label='degree 3')
plt.plot(x,Predict.Degree4,linewidth='3',color='black',linestyle='dashed',label='degree 4')
plt.legend(loc = 'best')
plt.title('Salary vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
