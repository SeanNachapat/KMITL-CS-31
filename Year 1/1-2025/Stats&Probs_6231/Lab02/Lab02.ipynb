```python
import numpy as np
import pandas as pd
import scipy.stats as stats
df = pd.read_csv('BloodPressure.csv')
df
```

```python
df_male = df.loc[df["sex"] == "Male"]
```

```python
df_female = df.loc[df["sex"] == "Female"]
```

## **T-Test**

```python
significance = 0.05
```

```python
stat, p_value = stats.ttest_rel(a = df_male['bp_before'], b = df_male['bp_after'])
```

```python
print('t_stat: %0.5f, p_value: %0.4f' % (stat, p_value))
if p_value < significance:
  print ("Reject the null hypothesis")
else:
  print ("Accept the null hypothesis")
```

```python
stat, p_value = stats.ttest_rel(a = df_female['bp_before'], b = df_female['bp_after'])
```

```python
print('t_stat: %0.5f, p_value: %0.4f' % (stat, p_value))
if p_value < significance:
  print ("Reject the null hypothesis")
else:
  print ("Accept the null hypothesis")
```
